## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import numpy as np
import random

def roll_dice():
    return random.randint(1,6)

wanted_result = 0
for _ in range(1000):
    pair = (roll_dice(),roll_dice())
    result = np.sum(pair)
    #print(f"par: {pair}")
    #print(f"suma del par: {result}")
    if result > 7 or result % 2 == 0:
        wanted_result += 1
    #print(f"conteo resultados favorables:{wanted_result}")
print(f"Probabilidad empírica: {wanted_result/1000} o {(wanted_result/1000)*100}%")
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

            
# select random ball function:
def select_random_ball(dict):
    random_key = random.choice(list(dict.keys()))
    chosen = dict[random_key]
    return chosen

# select 3 random balls using previews function:
def select_5_random_balls():
    selected_balls = []
    for _ in range(5):
        ball = select_random_ball(ball_box)
        selected_balls.append(ball)
    return selected_balls

#perform simulation from instructions:
three_white_two_red = 0
all_the_same = 0
for _ in range(1000):   
    white = 0
    red = 0
    green = 0
    selection = select_5_random_balls()
    for element in selection:
        if element == "White":
            white+=1
        elif element == "Red":
            red+=1
        else:
            green+=1
    if white == 3 and red == 2:
        three_white_two_red+=1
    elif white==5 or red==5 or green==5:
        all_the_same+=1

print(f"3 white balls and 2 red balls: {three_white_two_red} times out of 1000")
print(f"Probability of getting 3 white balls and 2 red balls: {three_white_two_red/1000} or {(three_white_two_red/1000)*100}%")
print(f"All the same color: {all_the_same} times out of 1000")
print(f"Probability all the same color: {all_the_same/1000} or {(all_the_same/1000)*100}%")
```
